```python
def fetch_director(obj):
    L=[]  
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
playlist_df["artist_name"] = playlist_df['artist_name'].apply(fetch_director)
```
